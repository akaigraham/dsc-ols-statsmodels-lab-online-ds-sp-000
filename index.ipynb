## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
plt.style.use('ggplot')
df = pd.read_csv('advertising.csv', index_col=0)
```

```python
# Check the columns and first few rows
df.head()
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
# Describe the contents of this dataset
# there are 200 observations
# Advertising sources included in the dataset are TV, radio, newspaper
# there is also a sales volume column
```

## Step 2: Plot histograms with kde overlay to check the distribution of the predictors

```python
# For all the variables, check distribution by creating a histogram with kde
plt.figure()
sns.distplot(df['TV'])
plt.title('TV Dist Plot')

plt.figure()
sns.distplot(df['radio'])
plt.title('Radio Dist Plot')

plt.figure()
sns.distplot(df['newspaper'])
plt.title('Newspaper Dist Plot')

plt.figure()
sns.distplot(df['sales'])
plt.figure('Sales Volume Dist Plot')
```

```python
# Record your observations here 
# TV and Radio both appear to have bimodal distributions 
# sales appears fairly normally distributed around ~10, but with some slight positive skew
# Newspaper has a slight positive skew
```


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
fig, axs = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(20, 5))
for index, channel in enumerate(list(df.columns[:3])):
    df.plot(kind='scatter', x=channel, y='sales', label=channel, ax=axs[index])
```

```python
# Record yor observations on linearity here 
# TV and Radio appear to have a linear, positive relationship with against sales volume
# newspaper does not seem to present much of a relationship
```

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols

# build the formula 
formula = 'sales ~ TV'

# create a fitted model in one line
model = ols(formula=formula, data=df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
new_X = pd.DataFrame({'TV': [df['TV'].min(), df['TV'].max()]})
print(new_X)

# make predictions for those x values and store them
predictions = model.predict(new_X)
print(predictions)

# first, plot the observed data and the least squares line
plt.figure(figsize=(8, 5))
plt.scatter(df['TV'], df['sales'])
plt.plot(new_X, predictions, color='blue')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Sales vs. TV')
plt.legend(['Least Squares Line', 'Observed Data'])
plt.show()
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model, 'TV', fig=fig)
plt.show()
```

```python
# Record Your observations on heteroscedasticity
# Looking at the residuals vs. TV, we see a slight cone shape, showing there is some heteroskedacity
# as a result, this violates our assumptions, and we cannot use OLS
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
plt.show()
```

```python
# Record Your observations on the normality assumption
# There does not appear to be a lot of deviation from the line, it appears this passes the normality assumption
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
# build the formula 
formula = 'sales ~ radio'

# create a fitted model in one line
model = ols(formula=formula, data=df).fit()
```

```python
model.summary()
```

```python
# create a DataFrame with the minimum and maximum values of radio
new_X = pd.DataFrame({'radio': [df['radio'].min(), df['radio'].max()]})
print(new_X)

# make predictions for those x values and store them
predictions = model.predict(new_X)
print(predictions)

# first, plot the observed data and the least squares line
plt.figure(figsize=(8, 5))
plt.scatter(df['radio'], df['sales'])
plt.plot(new_X, predictions, color='blue')
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.title('Sales vs. Radio')
plt.legend(['Least Squares Line', 'Observed Data'])
plt.show()
```

```python
# visualize error terms for variance and heteroskedacity
fig = plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model, 'radio', fig=fig)
plt.show()
```

Again, we can see that there appears to be some heteroskedacity here, looking at residuals versus radio - - as a result, this does not pass the homoskedacity assumption and is not a valid model

```python
# check the normality assumption 
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
plt.show()
```

Looking at the QQplot above, we see the distribution is not tight to the red line and violates the normality assumption

```python
# TV had a higher R-squared value in comparison to radio, but none of these variables is valid for 
# predicting sales due to violation of assumptions
```
